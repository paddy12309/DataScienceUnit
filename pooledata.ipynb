```python
import csv 
import pandas as pd
import matplotlib as mpl
from matplotlib import pyplot as plt
```

```python
with open('GlobalLandTemperaturesByCity.csv') as csvfile:
    data = pd.read_csv(csvfile)
    #wrangling some data to get only one city
    changedata = data[data['City'].str.match('Poole') & data['Country'].str.match('United Kingdom')]
    #changedata = changedata.set_index(changedata['dt'])
    #changedata = changedata['dt'].to_datetime()
    usedata = changedata[changedata['dt']>= '1900-01-01']
    print(usedata)
```
