```python
import numpy
from matplotlib import pyplot as plt
import pandas as pd
```

```python
with open('GlobalTemperatures.csv') as csvfile:
    data = pd.read_csv(csvfile)
    usedata = data[data['dt']>= '1900-01-01']
    usedata.dropna
print(usedata)
```

```python
x = usedata['dt']
y1 = usedata['LandAverageTemperature']
y2 = usedata['LandMaxTemperature']
y3 = usedata['LandMinTemperature']
plt.figure(figsize=(50,10))
plt.scatter(x,y1,label='average land temp')
plt.scatter(x,y2,label='max land temp') 
plt.scatter(x,y3,label='min land temp')
plt.legend()
plt.yscale('linear')
plt.gcf().autofmt_xdate()
plt.show()
```
