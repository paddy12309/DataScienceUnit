```python
import numpy
import csv
from matplotlib import pyplot as plt
import pandas as pd
```

```python
with open('GlobalTemperatures.csv') as csvfile:
    data = pd.read_csv(csvfile)
    data.dropna()
datatouselat = pd.DataFrame([])
datatouselmt = pd.DataFrame([])
b = 1900
for i in range(110):
    df = data[data['dt'].str.match(str(b))]
    sdf = data[data['dt'].str.match(str(b))]
#df = df.max()
    df = df.loc[df['LandAverageTemperature'].idxmax()]
    sdf = sdf.loc[sdf['LandMaxTemperature'].idxmax()]
    #datatouse = datatouse.append(df)
   # print(df)
    #print(b)
    b += 1
    #print(b)
    datatouselat = datatouselat.append(df)
    datatouselmt = datatouselmt.append(sdf)
#print(datatouselat)
#print(datatouselmt)
```

```python
x = datatouselat['dt']
y1 = datatouselat['LandAverageTemperature']
y2 = datatouselmt['LandMaxTemperature']
plt.figure(figsize=(30,20))
plt.plot(x,y1,label='average land temp')
plt.plot(x,y2,label='max land temp')
plt.legend()
plt.yscale('linear')
plt.gcf().autofmt_xdate()
plt.show()
```
